# Pipeline Building! 

## Lets cut the cheese one last time into our X and y values!

Final time for that joke as well! ;)

```python
X_train = train_df.drop(columns = 'FatLevel')
y_train = train_df['FatLevel']
X_test = test_df.drop(columns = 'FatLevel')
y_test = test_df['FatLevel']
```

Now let's identify the different types of feature data and split them into lists according to how we plan on transforming them.

First lets identify the numerical feature in our dataset.

```python
X_train.describe(include = 'all')
```

Looks like our two numerical features are **MoisturePercent** and **Organic**

```python
numerical_features = ['MoisturePercent','Organic']
```

Now let's identify our Categorical Features

Looks like we have **5** of them!

```python
categorical_features = ['ManufacturerProvCode','ManufacturingTypeEn','CategoryTypeEn','MilkTreatmentTypeEn','MilkTypeEn']
```

In this data we do not have any Binary (that needs to be transformed) or Ordinal Data

Let's start building our pipelines! These pipelines are important to being organized and making sure all our data is handled properly during transformation

```python
#Lets build the numerical feature pipeline
numeric_transformer = make_pipeline((SimpleImputer(strategy = 'median')),
                                    (StandardScaler()))

#Now the categorical feature pipeline
categorical_transformer = make_pipeline((SimpleImputer(strategy = 'most_frequent')),
                                        (OneHotEncoder(dtype = int ,  handle_unknown = "ignore")))
```

We can now combine these transformers into one transformer that runs them all at once.

```python
preprocessor = make_column_transformer((numeric_transformer, numerical_features),
                                      (categorical_transformer, categorical_features))
```

# Model Building!

We can start by creating a Dummy Model to act as a baseline!

```python
dummy = DummyClassifier(strategy = 'stratified')

#We can use cross validate and our dummy model to set a baseline for us.
dummy_scores = pd.DataFrame(cross_validate(dummy, X_train, y_train, cv = 5, return_train_score = True))

dummy_scores
```

We can see that our dummy model doesn't do too well in its scoring, lets get the mean values to compare to our actual models down the road!

```python
dummy_mean = dummy_scores.mean()
dummy_mean
```

# SVC Model

I chose SVC as one of my models, in case a high complexity solution is needed.

### Lets start off with the SVC model preparation:

Lets start off with an untuned model, to use as a baseline for this model!

We will create a pipeline for the SVC model using our **preprocessor** we made above!

```python
svc_pipe = make_pipeline(preprocessor, SVC(class_weight = 'balanced'))
```

Now we can run our pipeline through cross validation! 

```python
svc_score = cross_validate(svc_pipe, X_train, y_train, cv = 5, return_train_score = True)
svc_score = pd.DataFrame(svc_score)
svc_score.mean()
```

Looks like our model did better than our dummy classifier! Thats a good start lets tune those hyperparameters!

```python
#This will give values for our tuner to try!

svc_param_grid = {
    "svc__C": [0.01, 0.1, 1, 10, 100],
    "svc__gamma": [0.01, 0.1, 1, 10, 100]}

svc_rs = RandomizedSearchCV(svc_pipe,
                            svc_param_grid,
                            cv = 5,
                            n_iter = 10,
                            scoring = make_scorer(f1_score, pos_label = 'lower fat') , #using f1 score to judge the model, since we are not specifically looking for FN or FP as take aways
                            return_train_score = True, 
                            verbose = 1, 
                            n_jobs = -1,
                            )

svc_rs.fit(X_train, y_train)


svc_rs
```

Lets see what our best parameters and score are:

```python
svc_best_params = svc_rs.best_params_
svc_best_params
```

```python
svc_best_score = svc_rs.best_score_
svc_best_score
```

# KNeighbors Model

I am choosing to use KNeighbors as one of my models, due to the fact that fat levels should have a specific process in the making of cheese would lead to similar values

### Lets start off with the KNeighbors model preparation:

Lets start off with an untuned model, to use as a baseline for this model!

```python
knn_pipe = make_pipeline(preprocessor, KNeighborsClassifier())
```

Now we can run our pipeline through cross validation! 

```python
knn_score = pd.DataFrame(cross_validate(knn_pipe,
                                        X_train,
                                        y_train,
                                        cv = 5,
                                        return_train_score = True))
knn_score.mean()
```

Time to tune the hyperparameters!

```python
knn_params = {"kneighborsclassifier__n_neighbors":np.arange(1, 50, 5),
              "kneighborsclassifier__weights":['uniform','distance']}
knn_gs = GridSearchCV(knn_pipe, #We use GridSearchCV here since we are using set values for our parameters
                      knn_params,
                      cv = 5,
                      return_train_score =True,
                      n_jobs = -1,
                      scoring = make_scorer(f1_score, pos_label = 'lower fat')
                     )

knn_gs.fit(X_train, y_train)

knn_gs
```

```python
knn_best_params = knn_gs.best_params_
knn_best_params
```

```python
knn_best_score = knn_gs.best_score_
knn_best_score
```

# Logistic Regression

I want to use Logistic Regression as my final model, to gain insight on which feature is causing the largest impact!

### Lets start off with the KNeighbors model preparation:

Lets start off with an untuned model, to use as a baseline for this model!

```python
logr_pipe = make_pipeline(preprocessor, LogisticRegression(class_weight = 'balanced', max_iter = 2000))
```

Now we can run our pipeline through cross validation! 

```python
logr_score = pd.DataFrame(cross_validate(logr_pipe, X_train, y_train, cv = 5, return_train_score = True))
logr_score.mean()
```

Time to tune up the hyperparameters for our final model!

```python
logr_params = {"logisticregression__C": loguniform(0.01, 100)}
logr_rs =  RandomizedSearchCV(logr_pipe,
                              logr_params, 
                              cv = 5,
                              return_train_score =True,
                              n_jobs = -1,
                              scoring = make_scorer(f1_score, pos_label = 'lower fat'))
                   
                        
                        
logr_rs.fit(X_train, y_train)

logr_rs
```

```python
logr_best_params = logr_rs.best_params_
logr_best_params
```

```python
logr_best_score = logr_rs.best_score_
logr_best_score
```

### Since we used Logistic Regression we can check the coefficients of the features, to see their importance. Lets do this now!
First lets grab our best model, and the coefficients from that model, and the feature names

```python
logr_pipe.named_steps
```

```python
best_model = logr_rs.best_estimator_

logr_coeffs = best_model['logisticregression'].coef_

features = best_model.named_steps['columntransformer'].transformers_[1][1]\
                    .named_steps['onehotencoder'].get_feature_names(categorical_features)

#Then usingg the code from Assignment 8:

features_coef_df = pd.DataFrame(data = [features,logr_coeffs.flatten()]).T.rename(columns={0:'feature', 1:'coefficient'})
features_coef_df.sort_values(by = ['coefficient'], ascending = False)
```

Awesome! From this table of coefficients and labels we can see exactly which feature holds the most weight when descipering our label!
